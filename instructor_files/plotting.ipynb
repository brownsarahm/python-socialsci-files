# Plotting

```python
import pandas as pd
%matplotlib inline
```

```python
safi_df = pd.read_csv('data/SAFI_full_shortname.csv')
```

```python
# safi_df.columns
```

```python
safi_df['years_liv'].hist();
```

```python
safi_df['years_liv'].hist(bins=20)
```

```python
safi_df.hist(column='years_liv',by='village',
             layout=(1,3),figsize=(12,3),sharex=True,
            sharey=True);
```

```python
safi_df.hist(column='years_liv',by='village',
             layout=(1,3),figsize=(12,3),sharex=True,
            sharey=True);
```

```python
safi_df.plot.scatter(x='gps_Latitude',y='gps_Longitude',
                     c='gps_Altitude',colormap='viridis');
```

```python
%load code/plots
```

```python
# Make a scatter plot of 'years_farm' vs 'years_liv' and
# color the points by 'buildings_in_compound'
safi_df.plot.scatter(y='years_farm',x='years_liv',
                     c='buildings_in_compound',
                    colormap='viridis');
```

```python
# Make a bar plot of the mean number of rooms per wall type
safi_by_wall = safi_df.groupby('respondent_wall_type')

safi_by_wall['rooms'].mean().plot.bar();
```

```python
# Try plotting mean rooms per wall and roof type pairs
safi_by_wall_roof = safi_df.groupby(['respondent_wall_type',
                                     'respondent_roof_type'])

safi_by_wall_roof['rooms'].mean().plot.bar();
```

```python
safi_df.boxplot(by='village',column=['buildings_in_compound'])
```

```python
import seaborn as sns
sns.boxplot(data=safi_df,
            x='village',
            y='buildings_in_compound')
```

```python
import matplotlib.pylab as plt
```

```python
sns.boxplot(data=safi_df,
            x='village',
            y='buildings_in_compound')
plt.ylabel('Buildings in Compound')
plt.title('Building Number by Village')
plt.axis([-2,4,-1,10])
plt.savefig('buildings.png')
```

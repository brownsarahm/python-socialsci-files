# Extracting rows and columns from DataFrames

```python
# import pandas library
import pandas as pd
```

```python
# import the data
safi = pd.read_csv('data/SAFI_clean.csv')
```

```python
# check whether data are loaded correctly by printing first few rows
safi.head()
```

```python
# extracting the column 'village'
safi['village']
```

```python
# alternative method for extracting the column 'village'
safi.village
```

```python
# extracting 2 columns by putting a list [...,...] within the []
safi[['village', 'rooms']]
```

```python
# extracting rows 0 to 4, but not including 4 (remember python starts counting at 0!)
safi[0:4]
```

```python
# using iloc (integer based location) to extract rows 0 to 4 not including 4 (i.e. 0,1,2,3) and columns 1 to 4 not including 4 (1,2,3)
safi.iloc[0:4,1:4]
```

```python
# using loc to extract rows and columns by name, the name of the rows is the index, 
# the index is the very first column in safi, it does not have a name and is not included in the count of columns!
# using loc returns all rows/columns that are asked for, because here the numbers 0:4 refer to the name of the rows,
# i.e. the index and not the integer based position like with iloc
safi.loc[0:4, ['village','no_membrs']]
```

```python
# extracting all rows and column 'village'
safi.loc[:,'village']
```

### Solutions to exercise

```python
# 1. columns will be returned in the order they are listed
print(safi[['no_membrs', 'village']])
```

```python
# 2. column listed twice will be returned twice
print(safi[['no_membrs', 'village', 'no_membrs']])
```

```python
# 3. if you use a non existing column you will get an error message --> it says 'KeyError'
print(safi[['no_members', 'village']])
```

```python
# 4.
print(safi.iloc[:,2:5])
```

```python
# 5. You get an error if you only specify â€˜1â€™. You need to use â€˜:1â€™ or â€˜0:1â€™ to get the first row returned. 
# The â€˜:â€™ is always required. You can use â€˜:â€™ by itself to return all of the rows
print(safi[0:1]) 
```

```python
# 5. an alternative is to iloc, then you can use 0 for the first row
print(safi.iloc[0])
```

```python
# 6. you can use the function len(safi) to return the last row or tail()
safi[len(safi)-3:len(safi)]
```

```python
# 6. using tail
safi.tail(3)
```

## Conditional subsetting
using conditions, such as all rows where villag is Ruaca or all rows where rooms are more than (>) 3 ...

```python
# only return rows where village is Ruaca
ruaca = safi[(safi['village'] == 'Ruaca')]
```

```python
ruaca
```

```python
ruaca.shape
```

```python
# combine two conditions --> return rows where village is Ruaca and no_membrs are more than 15
safi[(safi['village'] == 'Ruaca') & (safi['no_membrs'] > 15)]
```

```python
# combine the previous selection with selection of columns by adding the 'command' for 
# selecting columns ([[...,...]]) at the end
safi[(safi['village'] == 'Ruaca') & (safi['no_membrs'] > 15)][['rooms','respondent_wall_type']]
```

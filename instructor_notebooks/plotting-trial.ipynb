# Plotting

```python
import pandas as pd
```

```python
%matplotlib inline
```

```python
safi_df = pd.read_csv("data/SAFI_full_shortname.csv")
safi_df.head()
```

```python
safi_df.dtypes
```

```python
safi_df['years_liv'].hist()
```

```python
safi_df['years_liv'].hist(bins=20)
```

```python
safi_df.hist(column='years_liv',by='village',layout =(1,3), figsize=(12,3),sharex=True)
```

```python
safi_df.plot.scatter(x='gps_Latitude', y='gps_Longitude', c='gps_Altitude', colormap="viridis", figsize=[10,10])
```

```python
safi_df.plot.scatter(x='years_farm', y='years_liv', c='buildings_in_compound', colormap="viridis", figsize=[10,10]);
```

```python
safi_df.boxplot(by ='village',column=['buildings_in_compound'])
```

```python
import seaborn as sns
```

```python
sns.boxplot(data=safi_df,x ='village',y='buildings_in_compound')
```

```python
sns.lmplot(x='years_farm', y='years_liv',data=safi_df,hue='village')
```

```python
sns.lmplot(x='no_plots', y='liv_count',data=safi_df,hue='village',x_jitter = .2,y_jitter=.2)
```
